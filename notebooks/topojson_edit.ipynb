```python
def remove_arc(topology, index_to_remove):
    if not isinstance(topology, tp.Topology):
        print("Error: Input must be a tp.Topology object.")
        return None

    topojson_data = topology.to_dict()
    arcs = topojson_data["arcs"]
    geometries = topojson_data["objects"][list(topojson_data["objects"].keys())[0]]["geometries"]

    if not (0 <= index_to_remove < len(arcs)):
        print(f"Error: Arc index {index_to_remove} is out of range.")
        return topology

    del arcs[index_to_remove]

    def edit_arc_ref(arc_ref):
        if abs(arc_ref) == index_to_remove:
            return None  # Signal to remove the geometry
        elif abs(arc_ref) > index_to_remove:
            return (arc_ref - 1) if arc_ref > 0 else (arc_ref + 1)
        else:
            return arc_ref

    geometries_to_remove = []

    for geom_index, geometry in enumerate(geometries):
        if 'arcs' in geometry:
            updated_arcs = []
            if isinstance(geometry['arcs'], list):
                if isinstance(geometry['arcs'][0], list):
                    for arc_list in geometry['arcs']:
                        updated_arc_list = []
                        for arc_ref in arc_list:
                            if isinstance(arc_ref, list):
                                updated_nested_list = []
                                for nested_ref in arc_ref:
                                    edited_ref = edit_arc_ref(nested_ref)
                                    if edited_ref is None:
                                        geometries_to_remove.append(geom_index)
                                        break
                                    updated_nested_list.append(edited_ref)
                                if geom_index in geometries_to_remove:
                                    break
                                updated_arc_list.append(updated_nested_list)
                            else:
                                edited_ref = edit_arc_ref(arc_ref)
                                if edited_ref is None:
                                    geometries_to_remove.append(geom_index)
                                    break
                                updated_arc_list.append(edited_ref)
                        if geom_index in geometries_to_remove:
                            break
                        updated_arcs.append(updated_arc_list)
                else: # single list of arcs
                    for arc_ref in geometry['arcs']:
                        edited_ref = edit_arc_ref(arc_ref)
                        if edited_ref is None:
                            geometries_to_remove.append(geom_index)
                            break
                        updated_arcs.append(edited_ref)
            else: # single arc reference.
                edited_ref = edit_arc_ref(geometry['arcs'])
                if edited_ref is None:
                    geometries_to_remove.append(geom_index)
                else:
                    updated_arcs = edited_ref

            if geom_index not in geometries_to_remove:
                geometry['arcs'] = updated_arcs

    geometries_to_remove = sorted(list(set(geometries_to_remove)), reverse=True)
    for index in geometries_to_remove:
        del geometries[index]

    topojson_data["arcs"] = arcs
    # topojson_data["objects"][list(topojson_data["objects"].keys())[0]]["geometries"] = geometries
    return tp.Topology(topojson_data)

topology_new = remove_arc(topology_old,index_to_remove=11)
display_arcs(topology_new)
print(topology_new)
```
